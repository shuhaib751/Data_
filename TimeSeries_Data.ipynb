```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data={
    "Date": pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01","2022-10-01","2022-11-01","2022-12-01","2023-01-01","2023-02-01","2023-03-01"]),
    "Value": [10,15,12,16,18,20,26,28,30,32,36,24,22,36,38]
}
```

```python
df=pd.DataFrame(Data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"])
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with seasonality")
plt.grid(True)
plt.show()
```

```python
data={
    "date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01"]),
    "value":[10,15,22,28,35]
}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='o')
plt.xlabel('Date')
plt.ylabel("Value")
plt.title("Time series with trend")
plt.grid(True)
plt.show()
```
