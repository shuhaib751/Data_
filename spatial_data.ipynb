```python
import folium

# Define the location using latitude and longitude
location = [13.169246990000547, 77.5348567563776]

# Create the map with specified parameters
mymap = folium.Map(location=location,
                   width=950,
                   height=550,
                   zoom_start=12,
                   tiles='openstreetmap')

# Add a marker at the specified location
folium.Marker(
    location=location,
    popup='Presidency University, Bangalore',  # Popup message when marker clicked
    icon=folium.Icon(color='blue')  # Marker color
).add_to(mymap)

# Save the map to an HTML file
mymap.save('presidency_university_map.html')

# Display the map in Jupyter Notebook
mymap
```

```python
import folium
from folium import features

location = [13.169246990000547, 77.5348567563776]


mymap = folium.Map(location=location,
                   width=950,
                   height=550,
                   zoom_start=12,
                   tiles='openstreetmap')

# Add a marker for Presidency University
folium.Marker(
    location=location,
    popup='Presidency University, Bangalore',  # Popup message when marker clicked
    icon=folium.Icon(color='blue')  # Marker color
).add_to(mymap)

# Add a custom icon for the marker
icon_url = 'https://image.flaticon.com/icons/png/512/65/65616.png'
icon = features.CustomIcon(icon_url, icon_size=(30, 30))
folium.Marker(
    location=location,
    popup='Presidency University, Bangalore',  # Popup message when marker clicked
    icon=icon
).add_to(mymap)

# Add pictures to the map
picture_urls = [
    'https://upload.wikimedia.org/wikipedia/commons/thumb/1/1a/Bangalore_palace.JPG/800px-Bangalore_palace.JPG',
    'https://www.example.com/image2.jpg'
]

for pic_url in picture_urls:
    iframe = folium.IFrame('<img src="' + pic_url + '" width="300" height="200">', width=350, height=250)
    popup = folium.Popup(iframe, max_width=400)
    folium.Marker(location=location, popup=popup).add_to(mymap)

# Save the map to an HTML file
mymap.save('presidency_university_map.html')

# Display the map in Jupyter Notebook
mymap
```
